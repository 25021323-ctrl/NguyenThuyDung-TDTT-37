1.Cước phí Taxi Thông Minh. 
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, kiểu số nguyên, đơn vị VND).

```python
import math
def smart_taxi_fare(car_type,start_hour,km,wait_minutes):
    fare = 0.0
    if car_type == 4:
        if km<= 0.8:
            fare = 11000
        elif 0.8 < km <= 20:
            fare = 11000 + 13000 * (km - 0.8)
        else:
            fare = 11000 + 13000 * 19.2 + 10500 * (km - 20)
    elif car_type == 7:
        if km <= 0.8:
            fare = 12000
        elif 0.8 < km <= 20:
            fare = 12000 + 14000 *(km-0.8)
        else:
            fare = 12000 + 14000 * 19.2 + 12000 * (km - 20)
    wait_fee = 0.0
    if 0 <= wait_minutes <= 5:
        wait_fee = 0
    else:
        wait_fee = math.ceil((wait_minutes - 5) / 5) * 2000
    fare += wait_fee
    night = start_hour in [22,23,0,1,2,3,4,5]
    if night:
        fare *= 1.2
    if km > 30 and not night:
        fare *= 0.9
    if fare < 50000 and km >5:
        fare = 50000
    return int(fare)
car_type, start_hour, km ,wait_minutes = map(float,input().split())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))

```

14.Viết hàm trả về hệ số nhị thức C(n, k). 

```python
import math
def nhi_thuc(n,k):
    if k<0 or k>n:
        return 0
    tu = math.factorial(n) # giai thừa
    mau = math.factorial(k) * math.factorial(n-k)
    return tu // mau
n,k = map(int,input().split())
print(nhi_thuc(n,k))
```

```python
def nhi_thuc(n,k):
    if k<0 or k>n:
        return 0
    tu = 1
    mau1 = 1
    mau2 =1
    for i in range(1,n+1):
        tu *= i
    for a in range(1,k+1):
        mau1 *= a
    for m in range(1,n-k+1):
        mau2 *= m
    return tu // (mau1*mau2)
n,k = map(int,input().split())
print(nhi_thuc(n,k))
    
```

25.Viết hàm point_in_quadrilateral(quad, p) trả về True nếu điểm p nằm bên trong hoặc trên biên của tứ giác quad, 
và False nếu ngược lại. - Đầu vào: - quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm, 
nhưng không tự cắt). - p: một điểm (x, y). - Đầu ra: giá trị boolean. - Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm 
trên biên là ở trong. 

```python
def point_in_quadrilateral(quad,p):
    px, py = p
    n = len(quad)
    inside = False
    for i in range(n):
        x1,y1 = quad[i]
        x2,y2 = quad[i-1]
        if min(y1,y2) <= py <= max(y1,y2) and min(x1,x2) <= px <= max(x1,x2):
            if (y2-y1)*(px - x1) == (x2-x1) * (py-y1):
                return True
        if ((y1>py) != (y2>py)) and (px < (x2-x1) * (py - y1)/(y2-y1)+x1):
            inside = not inside
    return inside
quad_input = []
for i in range(4):
    x,y = map(float,input().split())
    quad_input.append((x,y))
px = float(input())
py = float(input())
p_test = (px,py)
print(point_in_quadrilateral(quad_input,p_test))
```

30. Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu chấm động, ngược lại trả về 'neither'. 
- Đầu vào: chuỗi s. 
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. 
- Quy tắc: 
- Cho phép có dấu + hoặc - ở đầu. 
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). 
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). 
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này

```python
def detect_number_type(s):
    temp = s.lstrip('+-') # loại bỏ dấu ở đầu
    if temp.isdigit():
        return 'int'
    parts = temp.split('.') # Cắt chuỗi làm đôi tại dấu chấm
    if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
        return 'float'
    return 'neither'
s = input()
print(detect_number_type(s))
```

```python
import math
def is_right_triangle(a,b,c):
    if a + b > c or a+c > b or b+c > a:
        if math.isclose(a**2 + b**2, c**2) or math.isclose(a**2 +c**2, b**2) or math.isclose(b**2 + c**2, a**2):
            return True
    return False
a,b,c = map(float,input().split())
print(is_right_triangle(a,b,c))
  
```

33.Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op. 
- Đầu vào: các số a, b và ký tự op thuộc {'+', '-', '', '/'}. 
- Đầu ra: kết quả số học. 
- Yêu cầu thêm: 
- Nếu op là '/' và b == 0, phát sinh ValueError. 
- Nếu op không hợp lệ, phát sinh ValueError.

```python
def compute(a,b,op):
    if op == '+':
        return a + b
    elif op == '-':
        return a-b
    elif op == '*':
        return a*b
    elif op == '/':
        if b == 0:
            raise ValueError("Nhập b khác 0")
        return a/b
    else:
        raise ValueError
user_input = input().split() # nhập 5 + 3
a = float(user_input[0])
b = float(user_input[2])
op = user_input[1]
print(compute(a,b,op))
```

43.Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể 
tạo thành tam giác. 

```python
def tam_giac(a,b,c):
    if a + b > c or a + c > b or b + c > a:
        p = (a + b + c)/2
        S = (p*(p-a)*(p-b)*(p-c))**0.5
        return S
    raise ValueError('Ba cạnh không thể tạo thành tam giác')
a,b,c = map(int,input().split())
print(tam_giac(a,b,c))
```
