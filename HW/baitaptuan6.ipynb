1.Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. 
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3] 

```python
a = list(map(int,input().split()))
n = set()
ket_qua = []
for so in a:
    if so not in n:
        n.add(so)
        ket_qua.append(so)
print(ket_qua)
```

3.Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple 
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp 
xếp.) 

```python
a = tuple(map(int,input().split()))
k = int(input())
n = len(a)
if n>0:
    k = k%n #Nếu 5 ptu mà muốn xoay trái 7vtri thì giống xoay 2vtri
    ket_qua = a[k:] + a[:k] #[k:] lấy các ptu từ k đến hết + [:k] lấy từ đầu đến k
else:
    ket_qua = a
print(ket_qua)

    
```

4.Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp 
lại, thêm vào list của key đó 

```python
chuoi = input().split()
ket_qua = {}
for item in chuoi:
    key,value = item.split(':')
    if key in ket_qua:
        ket_qua[key].append(value)
    else:
        ket_qua[key] = [value]
print(ket_qua)
```

```python
chuoi = input().split()
ket_qua = {}
for item in chuoi:
    key, value = item.split(':')
    ket_qua.setdefault(key,[]).append(value)
print(ket_qua)
```

5.Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0) 

```python
a = list(map(int,input().split()))
ket_qua = {'positives': 0,'negatives' : 0, 'zeros' : 0}
for so in a:
    if so > 0:
        ket_qua['positives'] += 1
    elif so < 0:
        ket_qua['negatives'] += 1
    else:
        ket_qua['zeros'] += 1
print(ket_qua)
```

9.Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key.  

```python
import ast
dict1 = ast.literal_eval(input()) #để chuyển chuỗi nhập vào thành dict 1 cách an toàn
dict2 = ast.literal_eval(input()) #Nhập {'a':1}
dict_moi = dict1.copy()
for key, value in dict2.items():
    if key in dict_moi:
        dict_moi[key] += value #Nếu key đã có, cộng dồn giá trị
    else:
        dict_moi[key] = value
ket_qua = dict(sorted(dict_moi.items()))
print(ket_qua)
```

10.Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
a = list(map(int,input().split()))
k = int(input())
n = set() # Để lưu các số đã duyệt qua
ket_qua = set()
for so in a:
    phan_bu = k - so #Tính số phần bù cần tìm để tổng = k
    if phan_bu in n:
        ket_qua.add((phan_bu,so))
    n.add(so)
ket_qua_cuoi = sorted(list(ket_qua))
print(ket_qua_cuoi)
```

12.[Tìm phần tử xuất hiện nhiều nhất] 
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó. 

```python
a = list(map(int,input().split()))
dem_so = {}
for so in a:
    dem_so[so] = dem_so.get(so,0) + 1 # Đếm số lần xuất hiện của từng số
max_xuat_hien = max(dem_so.values()) #Tìm số lần xuất hiện nhiều nhất
cac_so_nhieu_nhat = [] #Lọc các tất cả các số đạt được số lần xuất hiện nhiều nhất
for so,so_lan in dem_so.items():
    if so_lan == max_xuat_hien:
        cac_so_nhieu_nhat.append(so)
print(min(cac_so_nhieu_nhat))
```

13.[Đảo ngược key và value của dictionary] 
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một 
dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In 
ra dictionary mới. 

```python
chuoi = input().split()
dao_nguoc = {}
for i in range(0,len(chuoi), 2):
    key_cu = chuoi[i]
    value_cu = chuoi[i+1]
    dao_nguoc[value_cu] = key_cu
print(dao_nguoc)
```

14.[Giao của hai danh sách] 
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các 
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các 
phần tử chung đó, không trùng lặp.

```python
a = list(map(int,input().split()))
b = list(map(int,input().split()))
moi = []
for i in a:
    if i in b and i not in moi:
        moi.append(i)
print(moi)
```

15.[Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key
value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới. 

```python
a = input().split()
k = int(input())
ket_qua ={}
for i in range(0,len(a),2):
    key = a[i]
    value = int(a[i+1])
    if value > k:
        ket_qua[key] = value
print(ket_qua)
        
```
