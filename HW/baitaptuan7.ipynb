1.Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target):

```python
def nhi_phan(a,n):
    low = 0 
    high = len(a) - 1
    while low <= high:
        mid = (low + high) // 2
        if a[mid] == n:
            return mid
        elif a[mid] < n:
            low = mid + 1
        else:
            high = mid - 1
    return -1
a = list(map(int,input().split()))
n = int(input())
print(nhi_phan(a,n))
```

3.Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection Sort) 

```python
def sap_xep(a):
    n = len(a)
    for i in range(n):
        for j in range(0,n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a
a = list(map(int,input().split()))
print(sap_xep(a))
```

4.Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất trong danh sách. Nếu có 
nhiều phần tử cùng số lần, trả về phần tử xuất hiện trước. 

```python
def xuat_hien(a):
    counts = {}
    for num in a:
        counts[num] = counts.get(num,0) + 1
    max_count = max(counts.values())
    for num in a:
        if counts[num] == max_count:
            return num, max_count
a = list(map(int,input().split()))
num, count = xuat_hien(a)
print(f"{num} xuat hien nhieu nhat, som nhat, {count} lan")
```

5.Cho danh sách các số nguyên và một số X, đếm tất cả các cặp (a, b) sao 
cho a + b = X. c 

```python
def cap(a,x):
    count = 0
    n = len(a)
    for i in range(n):
        for j in range(i+1,n):
            if a[i] + a[j] == x:
                count += 1
    return count
a = list(map(int,input().split()))
x = int(input())
print(cap(a,x))
```

6.Bạn được cung cấp một danh sách các danh sách lồng nhau, vd [[1, 2, 3], [4, 
5], [6, 7, 8, 9]], hãy làm phẳng danh sách

```python
def lam_phang(a):
    ket_qua = []
    for i in range(len(a)):
        if i % 2 != 0:
            ket_qua.extend(a[i][::-1])
        else:
            ket_qua.extend(a[i])
    return ket_qua
chuoi = input()
a = eval(chuoi)
print(lam_phang(a))
```

7.Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng 
nghiêm ngặt dài nhất (không nhất thiết phải liền kề).

```python
def do_dai(a):
    chuoi = []
    for i in a:
        left,right = 0 , len(chuoi) - 1 #tương đương left = 0, right = len(chuoi) - 1
        while left <= right:
            mid = (left + right) // 2
            if chuoi[mid] < i:
                left = mid + 1
            else:
                right = mid - 1
        if left == len(chuoi): #Nếu i lớn hơn tất cả các số trong chuỗi, thêm nó vào cuối
            chuoi.append(i)
        else:
            chuoi[left] = i
    return len(chuoi)
a = eval(input())
print(do_dai(a))
```

8.Bạn được cung cấp: một danh sách các khoảng thời gian interval = [[l1, r1], [l2, r2], ...] và một danh sách các truy vấn queries = [q1, q2, ...]  
Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của khoảng thời gian [li, ri] sao cho li ≤ q ≤ ri.  Nếu không tồn tại khoảng thời gian nào như vậy, hãy trả về -1 cho truy vấn đó. solve in < O(m.n), hint: binary search in the queries list 

```python
import heapq
def do_dai_nho_nhat(intervals, queries):
    m = len(queries)
    sorted_queries = sorted([(q,i) for i, q in enumerate(queries)])
    intervals.sort(key = lambda x: x[0])
    ans = [-1] * m
    min_heap = []
    i = 0
    n = len(intervals)
    for q, original_idx in sorted_queries:
        while i<n and intervals[i][0] <= q:
            l,r = intervals[i]
            length = r - l + 1
            heapq.heappush(min_heap, (length,r))
            i += 1
        while min_heap and min_heap[0][1] < q:
            heapq.heappop(min_heap)
        if min_heap:
            ans[original_idx] = min_heap[0][0]
    return ans
intervals = eval(input())
queries = eval(input())
print(do_dai_nho_nhat(intervals,queries))
```

9.Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có 
bao nhiêu phần tử bên phải nhỏ hơn nums[i].  Trả về danh sách các số đếm. 
Efficient solution: using heap or binary search tree 

```python
import bisect
def count(a):
    ans = []
    sorted_list = []
    for i in reversed(a): # duyệt ngược mảng từ phải sang trái
        idx = bisect.bisect_left(sorted_list,i)
        ans.append(idx)
        bisect.insort(sorted_list, i) # Chèn i vào 
    return ans[::-1]
a = eval(input())
print(count(a))

```

10.Cho một danh sách các số nguyên nums và một số nguyên m, hãy tìm tổng mảng con lớn nhất theo modulo m.  Tức là, với bất kỳ mảng con nums[i:j] nào, hãy tính sum(nums[i:j]) % m và trả về kết quả lớn nhất có thể. 
Gợi ý: Duy trì tổng tiền tố (prefix sum) mod m và sử dụng tìm kiếm nhị phân để tìm tổng tiền tố nhỏ nhất chỉ lớn hơn tổng hiện tại một chút (để tối đa hóa hiệu số bao quanh).

```python
import bisect
def tong_tien(nums,m):
    max_mod_sum = 0
    prefix_sum  = 0
    sorted_prefixes = [] # mảng lưu tổng tiền tố đẫ duyệt
    for num in nums:
        prefix_sum = (prefix_sum + num ) % m # Tính tổng tiền tố modulo m tại vị trí hiện tại
        max_mod_sum = max(max_mod_sum, prefix_sum) #Cập nhật ký lục nếu mảng con là lớn nhất
        idx = bisect.bisect_right(sorted_prefixes, prefix_sum) #bisect_right tìm ra index của phần tử đầu tiên thỏa mãn > prefix_sum
        if idx < len(sorted_prefixes):
            candidate = (prefix_sum - sorted_prefixes[idx] + m) % m
            max_mod_sum = max(max_mod_sum, candidate)
        bisect.insort(sorted_prefixes, prefix_sum) #Chèn PrefixSum hiện tại vào mảng lưu trữ sao cho vẫn giữ nguyên thứ tự sắp xếp
    return max_mod_sum
nums = eval(input())
m = int(input())
print(tong_tien(nums,m))
```

13.Dãy hài hòa là một dãy thỏa mãn điều kiện giá trị lớn nhất và giá trị nhỏ nhất của dãy khác nhau đúng bằng \( 1 \). Viết hàm trả về độ dài của dãy hài hòa con dài nhất có thể có trong mảng giá trị nguyên \( arr \) có \( n \) phần tử 
Ví dụ: Input: 1 3 2 2 5 2 3 7 
Output: 5 - Dãy hài hòa con dài nhất là [3,2,2,2,3]

```python
def hai_hoa(a):
    counts = {}
    for x in a:
        counts[x] = counts.get(x,0) + 1
    max_length = 0
    for x in counts:
        if x + 1 in counts:
            current_length = counts[x] + counts[x + 1]
            max_length = max(max_length, current_length)
    return max_length
a = list(map(int,input().split()))
print(hai_hoa(a))
```
