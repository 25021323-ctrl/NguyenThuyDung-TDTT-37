8.Yêu cầu: Cho chuỗi thời gian dạng "hh:mm:ss", hãy trả về tổng số giây.  
Giờ, phút, giây phải là số nguyên với 0 ≤ mm < 60 và 0 ≤ ss < 60.  
Trả về -1 nếu giá trị nằm ngoài phạm vi hợp lệ. 

```python
def time_to_total_seconds(a):
    parts = a.split(':')
    if len(parts) != 3:
        return -1
    hh = int(parts[0])
    mm = int(parts[1])
    ss = int(parts[2])
    if hh < 0 or mm < 0 or mm >= 60 or ss < 0 or ss >= 60:
        return -1
    seconds = hh * 3600 + mm * 60 + ss
    return seconds
a = input()
print(time_to_total_seconds(a))
```

```python
def rotate_right(a,x):
    n = len(a)
    if n == 0:
        return []
    x = x%n #TH x lớn hơn độ dài
    ket_qua = a[n-x:] + a[:n-x]
    return ket_qua
a = list(map(int,input().split()))
x = int(input())
print(rotate_right(a,x))
```

```python
def is_perfect(a):
    tong = 0
    if a<= 0:
        return False
    for i in range(1,a):
        if a % i == 0:
            tong += i
    if tong == a:
        return True
    return False
a = int(input())
print(is_perfect(a))
```

30.Cài đặt giải mã Atbash cho các chữ cái tiếng Anh. Ánh xạ a→z, b→y, ..., z→a và tương tự A→Z, ..., Z→A.  
Các ký tự không phải chữ cái được giữ nguyên. 

```python
def anh_xa(chuoi):
    ket_qua = []
    for i in chuoi:
        if 'a' <= i <= 'z':
            ky_tu_moi = chr(ord('z') - (ord(i) - ord('a')))
            ket_qua.append(ky_tu_moi)
        elif 'A' <= i <= 'Z':
            ky_tu_moi = chr(ord('Z') - (ord(i) - ord('A')))
            ket_qua.append(ky_tu_moi)
        else:
            ket_qua.append(i)
    return "".join(ket_qua)
chuoi = input()
print(anh_xa(chuoi))
```

Khai báo hàm: can_make_sum_odd(a) -> bool 

```python
def tong_le(a):
    n = len(a)
    count_odd = 0
    for i in a:
        if i % 2 != 0:
            count_odd += 1
    count_even = n - count_odd
    if count_odd == 0:
        return False
    if count_even == 0 and n % 2 == 0:
        return False
    return True
a = list(map(int,input().split()))
print(tong_le(a))
```

38.Viết hàm inverted_triangle(n) trả về một tam giác ngược cân bằng gồm các dấu sao *, có độ rộng là n. - n phải là một số nguyên dương lẻ, nếu không thì trả về chuỗi rỗng. - Mỗi dòng có đúng n ký tự, số dấu sao giảm dần 2 đơn vị mỗi dòng cho đến khi còn 1 dấu sao.

```python
def inverted_triangle(n):
    if n <= 0 or n % 2 == 0:
        return ""
    ket_qua = []
    for i in range(n,0,-2):
        spaces = (n-i) // 2
        dong = " " * spaces + "*" * i + " " * spaces
        ket_qua.append(dong)
    return "\n".join(ket_qua)
n = int(input())
print(inverted_triangle(n))        

```

44.Viết hàm count_solutions_3vars(a, b, c, t) trả về số nghiệm nguyên không âm (x, y, z) của phương trình a*x + b*y + c*z = t. Giả sử a, b, c, 
t là các số nguyên. Nếu a <= 0 hoặc b <= 0 hoặc c <= 0 hoặc t < 0 thì trả về 0.

```python
def count_solutions_3vars(a,b,c,t):
    if a <= 0 or b <= 0 or c <= 0 or t < 0:
        return 0
    so_nghiem = 0
    for x in range((t//a) + 1):
        phan_du_x = t - a * x
        for y in range((phan_du_x // b) + 1):
            phan_du_y = phan_du_x - b * y
            if phan_du_y % c == 0:
                so_nghiem += 1
    return so_nghiem
a,b,c,t = map(int,input().split())
print(count_solutions_3vars(a,b,c,t))
    
```

Khai báo hàm: top3_frequent_names(names) -> list[str]

47. Cho danh sách các bộ ba (tuple gồm 3 số nguyên), hãy trả về bộ ba xuất hiện nhiều nhất. 
Khai báo hàm: most_frequent_triple(triples) -> tuple[int,int,int]

```python
def most_frequent_triple(triples):
    if not triples:
        return None
    tan_suat = {}
    for t in triples:
        tan_suat[t] = tan_suat.get(t,0) + 1
    return max(tan_suat, key = tan_suat.get)
n = int(input())
danh_sach_triples = []
for i in range(n):
    dong = list(map(int,input().split()))
    bo_ba = tuple(dong)
    danh_sach_triples.append(bo_ba)
print(most_frequent_triple(danh_sach_triples))
```

48.Cho danh sách số nguyên nums và số nguyên k, hãy trả về True nếu tồn tại hai chỉ số khác nhau i và j sao cho: nums[i] == nums[j], và abs(i - j) <= k 
Khai báo hàm: contains_nearby_duplicate(nums, k) -> bool 

```python
def contains(nums,k):
    n = len(nums)
    for i in range(n):
        for j in range(i+1,n):
            if nums[i] == nums[j] and abs(i-j) <= k:
                return True
    return False
nums = list(map(int,input().split()))
k = int(input())
print(contains(nums,k))
```
