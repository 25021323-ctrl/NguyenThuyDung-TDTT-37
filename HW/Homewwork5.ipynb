16.Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x 
(Số Armstrong là một số nguyên dương mà tổng các chữ số của nó, với mỗi chữ số được lũy thừa với số lượng chữ số của chính nó, bằng đúng số ban đầu.)

```python
def amstrong(n):
    so_mu = len(str(n))
    tong = 0
    so_ban_dau = n
    while n>0:
        du = n % 10
        tong += du**so_mu
        n = n//10
    if tong == so_ban_dau:
        return True
    else:
        return False
n = int(input())
print(amstrong(n))

```

19. VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.
  số kỳ diệu là một số nguyên dương mà tổng các chữ số của nó, được cộng dồn liên tục cho đến khi chỉ còn một chữ số, có kết quả bằng 1.
VD : Xét số $1252$:Lần 1: $1 + 2 + 5 + 2 = 10$ (Vẫn còn 2 chữ số, tiếp tục cộng)Lần 2: $1 + 0 = 1$ ➔ Vì kết quả cuối cùng là 1, nên 1252 là số kỳ diệu.

```python
def so_ky_dieu(n):
    if n<= 0:
        return False
    while n > 9:
        tong = 0
        while n > 0:
            du = n % 10
            tong  += du
            n //= 10
        n = tong
    if tong == 1:
        return True
    else:
        return False
n = int(input())
print(so_ky_dieu(n))

```

20.Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'. 

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError: # Nếu bị lỗi, ta ép thành số thực
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
n = input()
print(detect_number_type(n))
```

24. Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không. 

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    tong = (x2-x1)*(y3-y1) - (y2-y1)*(x3-x1)
    if tong == 0:
        return True
    return False
x1,y1,x2,y2,x3,y3 = map(int,input().split())
print(are_collinear(x1,y1,x2,y2,x3,y3))
```

25.Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. 
Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad,p):
    px,py = p
    n = len(quad) #Số lượng đỉnh
    inside = False
    for i in range(n):
        x1,y1 = quad[i]
        x2,y2 = quad[i-1]
        if min(y1,y2) <= py <= max(y1,y2) and min(x1,x2) <= x2 <= max(x1,x2): # Kiểm tra điểm nằm trên cạnh
            if (y2-y1) * (px-x1) == (x2-x1)* (py-y1): # 3 điểm thẳng hàng khi hệ số góc bằng nhau
                return True
        if ((y1 > py) != (y2 > py)) and (px < (x2-x1)*(py-y1) / (y2-y1) + x1): #Kiểm tra nằm bên trong tứ giác
        # Vế 1 kt tia ngang đi qua P có nằm giữa 2 đầu y của cạnh không.Nếu có, tia này có khả năng cất cạnh
        # Vế 2 đây là công thức tìm hoành độ giao điểm của tia ngang với cạnh.Nếu P nằm bên trái giao điểm này nghĩa là tia đi từ P sang phải sẽ cắt cạnh
            inside = not inside
    return inside
quad_input = []
for i in range(4):
    x = float(input())
    y = float(input())
    quad_input.append((x,y))

px, py = map(float,input().split())
p_test = (px,py)
if point_in_quadrilateral(quad_input,p_test):
    print(f"{p_test} nằm trong hoặc trên biên tứ giác")
else:
    print(f"{p_test} nằm ngoài tứ giác")
```

28.Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def ham(x1,y1,x2,y2,x3,y3):
    tu = abs((x2-x1)*(x3-x1) + (y2-y1)*(y3-y1))
    mau1 = ((x2-x1)**2 + (y2-y1)**2)**0.5
    mau2 = ((x3-x1)**2 + (y3-y1)**2)**0.5
    goc = math.acos(tu/(mau1 * mau2))
    goc_do = math.degrees(goc)
    return round(goc_do,2)
x1,y1,x2,y2,x3,y3 = map(float,input().split())
print(ham(x1,y1,x2,y2,x3,y3))
```

34.Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau .Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    x = 0
    if n <= 0:
        return 0
    for i in range(1,n+1):
        s = str(i)
        if len(set(s)) == len(s):
            x += i
    return x
n = int(input())
print(sum_1_to_n(n))
```
