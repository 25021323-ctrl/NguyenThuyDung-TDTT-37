8.Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def doi_xung(so):
    chuoi = str(so)
    return chuoi == chuoi[::-1]
def dao_nguoc(so):
    return int(str(so)[::-1])
n = int(input())
so_buoc = 0
while not doi_xung(n):
    n = n + dao_nguoc(n)
    so_buoc += 1
print(f"{so_buoc} {n}")
```

10.Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất. 
(Tìm xem trong đoạn 1-n, số nào cần nhiều bước biến đổi nhất để về 1)

```python
def dem_buoc_collatz(x):
    so_buoc = 0
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3*x + 1
        so_buoc += 1
    return so_buoc
n = int(input())
max_buoc = -1
so_nho_nhat = 1
for i in range(1,n+1):
    buoc_hien_tai = dem_buoc_collatz(i)
    if buoc_hien_tai>max_buoc:
        max_buoc = buoc_hien_tai
        so_nho_nhat = i
print(f"{so_nho_nhat} {max_buoc}")
```

Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
def tong_uoc(n):
    tong = 0
    for i in range(1,int(n**0.5)+1):
        if n%i ==0:
            tong += i
            uoc_hai = n//i
            if uoc_hai != i:
                tong += uoc_hai
    return tong - n
a,b = map(int,input().split())
if tong_uoc(a) == b and tong_uoc(b) == a:
    print("True")
else:
    print("False")
```
