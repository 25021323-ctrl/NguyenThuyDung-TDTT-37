16.Yêu cầu 
Xây dựng lớp Shape với phương thức area(). 
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area(). 
Lấy π = 3.14. 
 
Input 
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE) 
Dòng 2: tham số tương ứng. 
 
Output 
In ra diện tích với 2 chữ số thập phân. 

```python
class Shape:
    def area(self): #area dùng để tính diện tích
        pass
class Rectangle(Shape):
    def __init__(self,w,h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h
class Circle(Shape):
    def __init__(self,r):
        self.r = r
    def area(self):
        return 3.14 * (self.r**2)
shape_type = input().strip()
params = list(map(float,input().split()))
if shape_type == 'RECTANGLE':
    shape = Rectangle(params[0],params[1]) #0 đóng vai trò chiều rộng, 1 đóng vai trò chiều dài
    print(f'{shape.area():.2f}')
elif shape_type == 'CIRCLE':
    shape = Circle(params[0])
    print(f"{shape.area():.2f}")
```

Dòng 1: loại đối tượng (Vehicle hoặc Car)

Xây dựng lớp Triangle(a, b, c) kế thừa từ Polygon và ghi đè perimeter() trả về a+b+c. 
 
Input 
Một dòng chứa 3 số nguyên a b c. 
 
Output 
In ra chu vi tam giác. 

```python
class Polygon:
    def perimeter(self):
        pass
class Triangle(Polygon):
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        return self.a + self.b + self.c
a,b,c = map(int,input().split())
triangle = Triangle(a,b,c)
print(triangle.perimeter())
```

20.Yêu cầu 
Xây dựng lớp Account(balance) với hai phương thức deposit(x) và get_balance(). 
Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức 
add_interest() để cộng balance*rate vào số dư. 
 
Input 
balance rate 
 
Output 
Số dư sau khi cộng lãi, in 2 chữ số thập phân. 
VD : 1000 0.05

```python
class Account:
    def __init__(self,balance):
        self.balance = balance
    def deposit(self,x): #Nạp tiền x vào tài khoản
        self.balance += x
    def get_balance(self): #Trả về số dư hiện tại
        return self.balance
class SavingAccount(Account):
    def __init__(self,balance,rate):
        super().__init__(balance)
        self.rate = rate
    def add_interest(self):
        self.balance += self.balance * self.rate
balance,rate = map(float,input().strip().split())
saving_acc = SavingAccount(balance,rate)
saving_acc.add_interest()
print(f"{saving_acc.get_balance():.2f}")
```

Xây dựng lớp BinaryPrinter kế thừa từ Printer, ghi đè print_value(x) để in biểu diễn nhị phân của 
số nguyên dương x. 
 
Input 
Một số nguyên dương n. 
 
Output 
In ra dạng nhị phân của n. 

```python
class Printer:
    def print_value(self,x):
        pass
class BinaryPinter(Printer):
    def print_value(self, x):
        binary_string = bin(x)[2:] #bin(x) chuyển x thành chuỗi nhị phân '0b1010'
        #[2:] dùng để cắt bỏ 2 ký tự ở đầu '0b'
        print(binary_string)
n = int(input().strip())
printer = BinaryPinter()
printer.print_value(n)
```

22.Yêu cầu 
Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận: 
[a b] 
[c d] 
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc. 
 
Input 
Một dòng chứa 4 số nguyên a b c d. 
 
Output 
In ra trace và determinant trên cùng một dòng.

```python
class Matrix:
    def __init__(self,a,b,c,d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return self.a + self.d
    def determinant(self):
        return self.a*self.d - self.b*self.c
a,b,c,d = map(float,input().split())
matrix = Matrix(a,b,c,d)
print(matrix.trace(),matrix.determinant())
```

Viết phương thức sum_str() trả về chuỗi biểu diễn đa thức tổng dưới dạng kx + m.

a b c d

Chuỗi của đa thức P(x)+Q(x).

27.Yêu cầu 
Xây dựng lớp Product(name, price). 
Xây dựng lớp DiscountProduct(name, price, percent) kế thừa từ Product, ghi đè phương thức 
final_price() để trả về giá sau giảm. 
Công thức: price * (100 - percent) / 100 
 
Input 
name price percent 
 
Output 
In ra giá cuối cùng với 2 chữ số thập phân.

```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price = price
    def final_price(self):
        pass
class Discount(Product):
    def __init__(self,name,price,percent):
        super().__init__(name,price)
        self.percent = percent
    def final_price(self):
        return self.price * (100 - self.percent) / 100
data = input().strip().split()
name = data[0]
price = float(data[1])
percent = float(data[2])
dis = Discount(name,price,percent)
print(dis.final_price()) 
```

28.Yêu cầu 
Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(), 
value(). 
Sau đó thực hiện dãy thao tác trên đối tượng. 
 
Input 
Dòng 1: n 
Dòng 2: n thao tác, mỗi thao tác là INC hoặc DEC, cách nhau bởi dấu cách. 
 
Output 
Giá trị cuối cùng của count.

```python
class Counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
    def dec(self):
        self.count -= 1
    def value(self):
        return self.count
n = int(input())
operations = input().strip().split() # ds cac thao tac
counter = Counter()
for op in operations:
    if op == 'INC':
        counter.inc()
    elif op == 'DEC':
        counter.dec()
print(counter.value())
```

Input 
h m s 
 
Output 
Tổng số giây.

```python
class Time:
    def __init__(self,h,m,s):
        self.h = h
        self.m = m
        self.s = s
    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s
h,m,s = map(int,input().split())
sec = Time(h,m,s)
print(sec.to_seconds())
```

30.Yêu cầu 
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x) tương đương với .append(x), pop() là .pop(), top() là lấy phần tử cuối qua chỉ số [-1] 
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ. 
 
Input 
Dòng 1: n 
n dòng tiếp theo là một trong ba lệnh: 
PUSH x 
POP 
TOP 
 
Output 
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng. 

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self,x):
        self.stack.append(x)
    def pop(self):
        self.stack.pop()
    def top(self):
        return self.stack[-1]
n = int(input().strip())
my_stack = Stack()
for _ in range(n):
    command = input().split()
    cmd_type = command[0]
    if cmd_type == 'PUSH':
        x = command[1]
        my_stack.push(x)
    elif cmd_type == 'POP':
        my_stack.pop()
    elif cmd_type == 'TOP':
        print(my_stack.top())



    
```
