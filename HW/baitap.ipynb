```python
n = int(input())
if n<= 0:
    print("Invalid")
else:
    print(f"{n} =", end=" ")
    temp = n
    d = 2
    res = []
    while d*d<=temp:
        if temp % d == 0:
            count = 0
            while temp % d == 0:
                count += 1
                temp //= d
            res.append(f"{d}^{count}")
        d += 1
    if temp > 1:
        res.append(f"{temp}^1")
    print("*".join(res))
```

```python
def uoc_so(n):
    tong_uoc = 0
    for i in range(1,int(n**0.5)+1):
        if n%i==0:
            tong_uoc += i
            uoc_hai = n // i
            if uoc_hai != i:
                tong_uoc += uoc_hai
    return tong_uoc - n
a,b = map(int,input().split())
if uoc_so(a) == b and uoc_so(b) == a:
    print("True")
else:
    print("False")
```
