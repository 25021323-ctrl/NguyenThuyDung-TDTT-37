16.Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 

```python
def hoan_hao(a):
    a.sort() #Sắp xếp mảng
    deletions = 0
    n = len(a)
    i = 0
    while i < n:
        count = 1
        current_val = a[i]
        while i + 1 < n and a[i+1] == current_val: #Đếm các số liên tiếp giống nhau
            count += 1
            i += 1
        if current_val == 0:
            deletions += count
        elif count > current_val:
            deletions += (count - current_val)
        elif count < current_val:
            deletions += count
        i += 1
    return deletions
a = eval(input())
print(hoan_hao(a))
```

17.Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ 
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. 
Nếu không có phần tử nào thỏa mãn thì không in gì 

```python
def xuat_hien(a):
    if not a:
        return 
    n = len(a)
    candidate = None
    count = 0
    for i in a:
        if count ==0 :
            candidate = i
        if i == candidate:
            count += 1
        else:
            count -= 1
    actual_count = 0
    for i in a:
        if i == candidate:
            actual_count += 1
    if actual_count > n //2 :
        return candidate
a = eval(input())
print(xuat_hien(a))
```

19.Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không 
dùng bất cứ thuật toán sắp xếp nào 

```python
def kiem_tra(a1,a2):
    if len(a1) > 1 and a1[0] > a1[-1]:#kiểm tra dãy giảm không
        a1 = a1[::-1]
    if len(a2) > 1 and a2[0] > a2[-1]:
        a2 = a2[::-1]
    ket_qua = []
    i , j = 0, 0
    len1, len2 = len(a1) , len(a2)
    while i < len1 and j < len2:
        if a1[i] <= a2[j] :
            ket_qua.append(a1[i])
            i += 1
        else:
            ket_qua.append(a2[j])
            j += 1
    while i < len1:
        ket_qua.append(a1[i])
        i += 1
    while j < len2 :
        ket_qua.append(a2[j])
        j += 1
    return ket_qua
a1 = eval(input())
a2 = eval(input())   
print(kiem_tra(a1,a2))
```

22.Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. 
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất 

```python
n = int(input()) #số lượng sinh viên
name_counts = {}
for _ in range(n):
    full_name = input().strip()
    words = full_name.split()
    if words:
        first_name = words[-1]
        name_counts[first_name] = name_counts.get(first_name,0) + 1 
best_name = ""
max_count = 0
for name, count in name_counts.items():
    if count > max_count:
        max_count = count
        best_name = name
    elif count == max_count:
        if best_name == "" or name < best_name:
            best_name = name
print(best_name)
```

25.Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi 
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số 

```python
def tim_so(s,k):
    stack = []
    so_luong_can_xoa = len(s) - k
    for digit in s:
        while stack and so_luong_can_xoa > 0 and stack[-1] < digit:
            stack.pop()
            so_luong_can_xoa -= 1
        stack.append(digit)
    if so_luong_can_xoa > 0:
        stack = stack[:-so_luong_can_xoa]
    return "".join(stack[:k])
chuoi_so = input()
k = int(input())
if k >= len(chuoi_so):
    print(chuoi_so)
elif k<= 0:
    print("Không có số nào")
else:
    print(tim_so(chuoi_so,k))
```

29.Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ 
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau 

```python
a = eval(input())
if not a: #Trường hợp danh sách trống
    print('Danh sách rỗng')
else:
    max_seq = []
    current_seq = [a[0]]
    for i in range(1,len(a)):
        if a[i] > a[i-1]:
            current_seq.append(a[i])
        else:
            if len(current_seq) > len(max_seq):
                max_seq = current_seq
            current_seq = [a[i]] # reset lại dãy
    if len(current_seq) > len(max_seq):
        max_seq = current_seq
    print(max_seq)
```

30.Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím 

```python
a = eval(input())
if not a:
    print('Danh sách trống')
else:
    n = len(a)
    # dp[i] lưu độ dài của dãy con tăng dài nhất kết thúc tại vị trí i
    dp = [1] * n #Ban đầu mỗi số tự tạo thành 1 dãy độ dài 1
    #trace[i] dùng để lưu dấu vết của số đứng ngay trước a[i] trong dãy
    trace = [-1] * n # -1 nghĩa là chưa có phần tử nào đứng trước
    for i in range(1,n):
        for j in range(i):
            if a[i] > a[j] and dp[i] < dp[j] + 1:
                dp[i] = dp[j] + 1
                trace[i] = j #phần từ a[j] đứng liền trước a[i]
    max_len = 0
    last_index = -1
    for i in range(n):
        if dp[i] > max_len:
            max_len = dp[i]
            last_index = i
    ket_qua = []
    while last_index != -1:
        ket_qua.append(a[last_index])
        last_index = trace[last_index]
    ket_qua.reverse()
    print(max_len)
    print(ket_qua)

```
