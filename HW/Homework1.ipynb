Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
a,b,c = map(int,input().split())
if a+b>c and a+c>b and b+c>a:
    p = (a+b+c)/2
    dien_tich= (p*(p-a)*(p-b)*(p-c))**0.5
    print(f"{dien_tich:.1f}")
else:
    print("Không phải 3 cạnh của tam giác")
```

9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 

```python
n = int(input())
match n:
    case 0:
        print("Không")
    case 1:
        print("Một")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bốn")
    case 5:
        print("Năm")
    case 6:
        print("Sáu")
    case 7:
        print("Bảy")
    case 8:
        print("Tám")
    case 9:
        print("Chín")
    case _:
        print("Không thỏa mãn")
```

```python
chu_so = {0: "Không", 1: "Một", 2: "Hai", 3: "Ba", 4: "Bốn", 5: "Năm", 6: "Sáu", 7: "Bảy", 8: "Tám", 9: "Chín"}
print(chu_so.get(int(input()), "Số không nằm trong từ 1 đến 9"))
```
