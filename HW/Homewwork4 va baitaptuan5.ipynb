Phép trừ \( - \)

10.Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai từ đầu vào có cùng độ dài.

```python
def dang_cau(a,b):
    if len(a) != len(b):
        return False
    a_to_b = {}
    b_to_a = {}
    for ky_tu_a, ky_tu_b in zip(a,b):
        if ky_tu_a in a_to_b:
           if a_to_b[ky_tu_a] != ky_tu_b:
                return False
        else:
            a_to_b[ky_tu_a] = ky_tu_b
        if ky_tu_b in b_to_a:
            if b_to_a[ky_tu_b] != ky_tu_a:
                return False
        else:
            b_to_a[ky_tu_b] = ky_tu_a
    return True
a,b = map(str,input().split())
print(dang_cau(a,b))
                
    
```

14.Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute(a,b,op):
    if op == '+':
        return a+b
    elif op == '-':
        return a-b
    elif op == '*':
        return a*b
    elif op == '/':
        return a/b
    else:
        return False
a,b = map(int,input().split())
c = compute(a,b,'+')
print(c)
```

15.Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def is_right_triangle(a,b,c):
    if a + b > c and a + c>b and b+c>a:
        if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
            return True
    return False
a,b,c = map(int,input().split())
print(is_right_triangle(a,b,c))
```
