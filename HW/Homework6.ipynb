16.Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải 

```python
n,m = map(int,input().split())
matrix = []
for i in range(n):# Nhập các phần tử của mảng
    row = [] #Mỗi lần bắt đầu 1 hàng mới
    for j in range(m):
        so = int(input(f"Nhập a[{i}][{j}] : "))
        row.append(so)
    matrix.append(row)
print("Ma trân vừa nhập là")
for i in range(n):
    for j in range(m):
        print(f"{matrix[i][j]:4d}", end=" ") #dành ra 4 khoảng trống và căn phải cho số nguyên
    print()
    
```

17.Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python
n = int(input())
matrix = []
for i in range(n):
    hang = list(map(int,input().split())) #Nhập theo từng hàng
    matrix.append(hang)
print("Đường chéo chính của ma trận là: ")
for i in range(n):
    print(f"{matrix[i][i]}", end=" ")
print()
print("Đường chéo phụ của ma trận là: ")
for i in range(n):
    print(f"{matrix[i][n-1-i]:4d}", end=" ")
print() #Xuống dòng
```

18.Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên 
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột 
k (hàng k) 

```python
n,m,k = map(int,input().split())
matrix = []
for i in range(n):
    hang = list(map(int,input().split()))
    matrix.append(hang)
if 0<= k < n:
    tong_hang = sum(matrix[k][j] for j in range(m)) # hàm sum() để tính tổng
    print(tong_hang)
else:
    print(f"Không tồn tại hàng{k} trong ma trận")
if 0<= k <m:
    tong_cot = sum(matrix[i][k] for i in range(n)) 
    print(tong_cot)
else:
    print(f"Không có cột{k} trong ma trận")

```

20.Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về

```python
def tim_so(a,k):
    n = len(a)
    for i in range(n):
        if a[i] == k:
            return i
    return -1
a = list(map(int,input().split()))
k = int(input())
print(tim_so(a,k))
```

21.Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

```python
a = input().split()
chuoi = {}
for so, tu in enumerate(a): # Vừa đếm stt, vừa lấy luộn giá trị
    if tu not in chuoi:
        chuoi[tu] = []
    chuoi[tu].append(so)
for tu in chuoi:
    chuoi[tu] = tuple(chuoi[tu])
print(chuoi)
```

24.Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9 

```python
def chuoi(a):
    ket_qua = []
    so_0 = 0
    for i in a:
        if i == 0:
            so_0 += 1
        else:
            ket_qua.append(i)
    return [0]*so_0 + ket_qua
a = list(map(int,input().split()))
print(chuoi(a))            
```

25.Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong mảng 

```python
def tim(a):
    day_du = set(range(10))
    tap_hop = set(a)
    so_thieu = day_du - tap_hop
    return sorted(list(so_thieu))
a = list(map(int,input().split()))
print(tim(a))
```

```python
def tim(a):
    ket_qua = []
    for i in range(10):
        if i not in a:
            ket_qua.append(i)
    return ket_qua
a = list(map(int,input().split()))
print(tim(a))
```

26.Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 
phần tử đầu và cuối 

```python
a = list(map(int,input().split()))
n = len(a)
tong = 0
for i in range(1,n-1):
    if a[i] > a[i-1] and a[i] > a[i+1]:
        tong += a[i]
print(tong)
```

27.Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1

28.Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
thực sự, hay vô danh (không tăng, không giảm) 

```python
a = list(map(int,input().split()))
n = len(a)
tang = True
giam = True
for i in range(n-1):
    if a[i+1] > a[i]:
        giam = False
    elif a[i+1]< a[i]: 
        tang = False
if tang and n > 1:
    print('Tăng')
elif giam and n > 1:
    print('Giảm')
else:
    print('Vô danh')
    
```
