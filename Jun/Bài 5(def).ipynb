```python
def tinh_giam_gia(gia_goc, phan_tram):
    so_tien_giam = gia_goc * (phan_tram /100)
    gia_cuoi = gia_goc - so_tien_giam
    return gia_cuoi
n = tinh_giam_gia(200000,15)
print(f"{n}")
   
```

Tạo hàm với 10 số nguyên tố

```python
def fun(n):
    x = 2
    count = 0
    while count<n:
        for d in range(2,int(x**0.5)+1):
            if x % d==0:
                break
        else:
            print(x)
            count+=1
        x +=1
n=10
fun(n)
```

```python
def incr_list(nums):
    for i in range(len(nums)):
        nums[i]= nums[i]+1
lst=[6,0,7]
incr_list(lst)
print(lst)
```

```python
def print_list(lst):
    for word in lst:
        print(word)
def print_haiku_with_helper(h):
    for line in h:
        print_list(line)
```
