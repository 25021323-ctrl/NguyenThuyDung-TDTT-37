Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu 
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần 
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 
chữ số phần thập phân.

```python
a = float(input("chiều rộng a: "))
b = float(input("chiều dài b: "))
dien_tich_hcn = a * b
ban_kinh = a / 2
dien_tich_hinh_tron = 3.14 * (ban_kinh **2)
dien_tich_con_lai = dien_tich_hcn - dien_tich_hinh_tron
print(f"Diện tích còn lại để trồng cây là: {dien_tich_con_lai:.2f}")
```

```python
c = input("Nhập kí tự c: ")
if c.isupper():
    ket_qua = c.lower()
    print(f"Chữ thường tương ứng là: {ket_qua}")
elif c.islower():
    ket_qua = c.upper()
    print(f"Chữ in hoa tương ứng là {ket_qua}")
else:
    print("kí tự bạn vừa nhập không phải là chữ cái!")
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
c = float(input("Nhập số c: "))
if (a+b>c) and (a+c>b) and (b+c>a):
    p = (a + b +c) / 2
    dien_tich = math.sqrt(p*(p-a)*(p-b)*(p-c)) 
    print(f"Diện tích của tam giác là: {dien_tich:.1f}")
else:
    print("Không phải 3 cạnh của tam giác")
```

Bài 9: Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...)

```python
so = int(input("Nhập một số nguyên từ 0 đến 9: "))
match so:
    case 0:
        print("không")
    case 1:
        print("một")
    case 2:
        print("hai")
    case 3:
        print("ba")
    case 4:
        print("bốn")
    case 5:
        print("năm")
    case 6:
        print("sáu")
    case 7:
        print("bảy")
    case 8:
        print("tám")
    case 9:
        print("chín")
    case _:
        print("Số nhập vào không hợp lệ")
```
